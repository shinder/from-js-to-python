對於熟悉 TypeScript 的 JavaScript 開發者來說，Python 的型別提示會感到非常親切。兩者都是在動態型別語言上添加靜態型別檢查的解決方案。

## 13.2 進階型別

### Optional - 可能為 None

````python
from typing import Optional

# Optional[X] 等同於 Union[X, None]
def find_user(user_id: int) -> Optional[str]:
    users = {1: "Alice", 2: "Bob"}
    return users.get(user_id)

print(f"find_user(1): {find_user(1)}")
print(f"find_user(3): {find_user(3)}")
````

````python
# Python 3.10+ 可以用更簡潔的語法
def find_user_v2(user_id: int) -> str | None:
    users = {1: "Alice", 2: "Bob"}
    return users.get(user_id)

print(f"find_user_v2(1): {find_user_v2(1)}")
print(f"find_user_v2(3): {find_user_v2(3)}")
````

### Union - 多種型別

````python
from typing import Union

# 接受多種型別
def process(data: Union[str, bytes]) -> str:
    if isinstance(data, bytes):
        return data.decode("utf-8")
    return data

print(process("hello"))
print(process(b"world"))
````

````python
# Python 3.10+ 語法
def process_v2(data: str | bytes) -> str:
    if isinstance(data, bytes):
        return data.decode("utf-8")
    return data

# 多於兩種型別
def flexible_input(value: int | float | str) -> str:
    return str(value)

print(flexible_input(42))
print(flexible_input(3.14))
print(flexible_input("hello"))
````

### Callable - 函式型別

````python
from typing import Callable

# 定義函式型別：接受參數型別和回傳型別
# Callable[[參數1型別, 參數2型別, ...], 回傳型別]
def apply_twice(func: Callable[[int], int], value: int) -> int:
    return func(func(value))

def double(x: int) -> int:
    return x * 2

result = apply_twice(double, 5)
print(f"apply_twice(double, 5) = {result}")
````

````python
# 更複雜的函式型別
Handler = Callable[[str, int], bool]

def process_with_handler(handler: Handler, name: str, count: int) -> bool:
    return handler(name, count)

def my_handler(name: str, count: int) -> bool:
    print(f"Processing {name} with count {count}")
    return count > 0

result = process_with_handler(my_handler, "test", 5)
print(f"Result: {result}")
````

````python
# 無參數函式
NoArgFunc = Callable[[], str]

def get_greeting() -> str:
    return "Hello!"

func: NoArgFunc = get_greeting
print(func())

# 任意參數
AnyArgsFunc = Callable[..., str]  # ... 表示任意參數

def flexible_func(*args, **kwargs) -> str:
    return f"args={args}, kwargs={kwargs}"

f: AnyArgsFunc = flexible_func
print(f(1, 2, name="test"))
````

### Any 與 object

````python
from typing import Any

# Any：任何型別，跳過型別檢查
def process_anything(data: Any) -> Any:
    return data

# object：所有型別的基底類別（比 Any 嚴格）
def print_object(obj: object) -> None:
    print(obj)

# 兩者的差異
print("Any 和 object 的差異：")
print("- Any：型別檢查器完全忽略，可以做任何操作")
print("- object：只能使用所有物件共有的方法")
````

### Literal - 字面值型別

````python
from typing import Literal

# 只接受特定值
def set_direction(direction: Literal["left", "right", "up", "down"]) -> None:
    print(f"Moving {direction}")

set_direction("left")   # OK
set_direction("right")  # OK

# 以下會在型別檢查時報錯（但執行時不會）
# set_direction("center")  # 型別錯誤！
````

### Final - 常數

````python
from typing import Final

# 宣告常數（不應被重新賦值）
MAX_SIZE: Final[int] = 100
API_URL: Final = "https://api.example.com"  # 型別會被推斷

print(f"MAX_SIZE: {MAX_SIZE}")
print(f"API_URL: {API_URL}")

# 類別中的常數
class Config:
    DEBUG: Final[bool] = False
    VERSION: Final = "1.0.0"

print(f"Config.DEBUG: {Config.DEBUG}")
print(f"Config.VERSION: {Config.VERSION}")
````

### TypeVar - 泛型型別變數

````python
from typing import TypeVar

# 定義泛型型別變數
T = TypeVar("T")

def first(items: list[T]) -> T:
    return items[0]

# 使用
print(first([1, 2, 3]))       # 回傳 int
print(first(["a", "b"]))     # 回傳 str
print(first([1.0, 2.0, 3.0])) # 回傳 float
````

````python
# 限制型別變數的範圍
Number = TypeVar("Number", int, float)

def add_numbers(a: Number, b: Number) -> Number:
    return a + b

print(add_numbers(1, 2))       # int
print(add_numbers(1.5, 2.5))   # float
````

````python
# 有上界的型別變數
from typing import Sequence

S = TypeVar("S", bound=Sequence)

def get_length(seq: S) -> int:
    return len(seq)

print(get_length([1, 2, 3]))     # list
print(get_length((1, 2, 3, 4)))  # tuple
print(get_length("hello"))       # str
````

### Generic - 泛型類別

````python
from typing import TypeVar, Generic

T = TypeVar("T")

class Stack(Generic[T]):
    def __init__(self) -> None:
        self._items: list[T] = []
    
    def push(self, item: T) -> None:
        self._items.append(item)
    
    def pop(self) -> T:
        return self._items.pop()
    
    def is_empty(self) -> bool:
        return len(self._items) == 0
    
    def __repr__(self) -> str:
        return f"Stack({self._items})"

# 使用
int_stack: Stack[int] = Stack()
int_stack.push(1)
int_stack.push(2)
int_stack.push(3)
print(f"int_stack: {int_stack}")
print(f"pop: {int_stack.pop()}")

str_stack: Stack[str] = Stack()
str_stack.push("hello")
str_stack.push("world")
print(f"str_stack: {str_stack}")
````

### Protocol - 結構化型別（鴨子型別）

Python 的 Protocol 類似 TypeScript 的 interface，實現結構化子型別。

````python
from typing import Protocol

# 定義 Protocol（類似 interface）
class Drawable(Protocol):
    def draw(self) -> None: ...

class Circle:
    def draw(self) -> None:
        print("Drawing circle")

class Square:
    def draw(self) -> None:
        print("Drawing square")

# 不需要顯式繼承，只要有相同的方法簽名即可
def render(shape: Drawable) -> None:
    shape.draw()

render(Circle())
render(Square())
````

````python
# 帶屬性的 Protocol
class Named(Protocol):
    name: str

class Person:
    def __init__(self, name: str):
        self.name = name

class Product:
    def __init__(self, name: str, price: float):
        self.name = name
        self.price = price

def greet(obj: Named) -> str:
    return f"Hello, {obj.name}"

print(greet(Person("Alice")))
print(greet(Product("Laptop", 999.99)))
````

### TypedDict - 型別化字典

````python
from typing import TypedDict

# 定義字典結構（類似 TypeScript 的 interface）
class UserDict(TypedDict):
    name: str
    age: int
    email: str

user: UserDict = {
    "name": "Alice",
    "age": 30,
    "email": "alice@example.com"
}

print(f"User: {user}")
print(f"Name: {user['name']}")
````

````python
# 混合必填和選填（Python 3.11+）
from typing import NotRequired

class UserDictV2(TypedDict):
    name: str                      # 必填
    age: int                       # 必填
    email: NotRequired[str]        # 選填
    nickname: NotRequired[str]     # 選填

# 只提供必填欄位
user1: UserDictV2 = {"name": "Bob", "age": 25}

# 提供所有欄位
user2: UserDictV2 = {
    "name": "Carol",
    "age": 28,
    "email": "carol@example.com",
    "nickname": "CC"
}

print(f"user1: {user1}")
print(f"user2: {user2}")
````

## 13.3 特殊型別工具

### cast - 型別轉換

````python
from typing import cast

# 告訴型別檢查器這是特定型別（不會在執行時轉換）
def get_config() -> dict:
    return {"debug": True, "version": "1.0"}

config = cast(dict[str, str | bool], get_config())
print(f"config: {config}")

# cast 只是給型別檢查器的提示，不會真的轉換
value = cast(int, "not an int")  # 不會報錯，但是危險！
print(f"value: {value}, type: {type(value)}")
````

### TYPE_CHECKING - 只在型別檢查時匯入

````python
from typing import TYPE_CHECKING

# TYPE_CHECKING 在執行時為 False，只在型別檢查時為 True
print(f"TYPE_CHECKING: {TYPE_CHECKING}")

# 常用於避免循環匯入
if TYPE_CHECKING:
    # 這個 import 只在型別檢查時執行
    from typing import List

# 使用字串形式的前向參照
class UserService:
    def get_user(self) -> "User":  # 使用字串
        pass

class User:
    pass
````

### 前向參照（Forward Reference）

````python
from __future__ import annotations  # 啟用延遲評估（Python 3.7+）

class Node:
    def __init__(self, value: int):
        self.value = value
        self.next: Node | None = None  # 自我參照
    
    def set_next(self, node: Node) -> None:
        self.next = node
    
    def __repr__(self) -> str:
        next_val = self.next.value if self.next else None
        return f"Node({self.value}, next={next_val})"

# 建立鏈結串列
n1 = Node(1)
n2 = Node(2)
n3 = Node(3)
n1.set_next(n2)
n2.set_next(n3)

print(n1)
print(n2)
print(n3)
````

### @overload - 函式重載

````python
from typing import overload

# 定義多個函式簽名
@overload
def process(x: int) -> int: ...

@overload
def process(x: str) -> str: ...

@overload
def process(x: list[int]) -> list[int]: ...

# 實際實作
def process(x: int | str | list[int]) -> int | str | list[int]:
    if isinstance(x, int):
        return x * 2
    elif isinstance(x, str):
        return x.upper()
    else:
        return [i * 2 for i in x]

print(process(5))           # 10
print(process("hello"))     # HELLO
print(process([1, 2, 3]))   # [2, 4, 6]
````

### TypeScript 對照

## 13.5 dataclass 與型別提示

````python
from dataclasses import dataclass, field
from typing import Optional

@dataclass
class User:
    name: str
    age: int
    email: Optional[str] = None
    tags: list[str] = field(default_factory=list)

# 自動生成 __init__, __repr__, __eq__ 等方法
user = User(name="Alice", age=30)
print(user)

user_with_email = User(name="Bob", age=25, email="bob@example.com", tags=["admin"])
print(user_with_email)
````

````python
# 不可變版本
@dataclass(frozen=True)
class Point:
    x: float
    y: float

p = Point(1.0, 2.0)
print(f"Point: {p}")

try:
    p.x = 3.0  # 會拋出 FrozenInstanceError
except Exception as e:
    print(f"錯誤：{type(e).__name__} - 不可變物件")
````

## 13.6 最佳實踐

### 1. 漸進式採用

不需要一次為所有程式碼加上型別，可以漸進式採用。

````python
from typing import Any

# 先從公開 API 開始
def public_api(data: dict[str, Any]) -> list[str]:
    # 內部實作可以暫時不加型別
    result = []
    for key, value in data.items():
        result.append(str(value))
    return result

print(public_api({"a": 1, "b": 2}))
````

### 2. 避免過度使用 Any

````python
from typing import TypeVar, Any

# 不好的做法
def process_bad(data: Any) -> Any:
    return data

# 好的做法
T = TypeVar("T")

def process_good(data: T) -> T:
    return data

# 使用泛型可以保留型別資訊
result = process_good(42)  # 型別檢查器知道結果是 int
````

### 3. 使用型別別名提高可讀性

````python
from typing import TypeAlias

# 型別別名
UserId: TypeAlias = int
UserName: TypeAlias = str
UserDict: TypeAlias = dict[UserId, UserName]

def get_users() -> UserDict:
    return {1: "Alice", 2: "Bob"}

users = get_users()
print(f"Users: {users}")
````

### 4. 使用 Protocol 而非具體類別

````python
from typing import Protocol

# 好的做法：使用 Protocol 定義介面
class Reader(Protocol):
    def read(self) -> str: ...

def process_input(reader: Reader) -> str:
    return reader.read().upper()

# 任何有 read() 方法的物件都可以使用
class StringReader:
    def __init__(self, content: str):
        self.content = content
    
    def read(self) -> str:
        return self.content

reader = StringReader("hello world")
print(process_input(reader))
````

### 5. 正確處理可變預設值

````python
from typing import Optional

# 不好的做法 - 可變預設值！
def bad_func(items: list[int] = []) -> list[int]:
    items.append(1)
    return items

# 問題示範
print(bad_func())  # [1]
print(bad_func())  # [1, 1] - 共用同一個 list！
print(bad_func())  # [1, 1, 1]
````

````python
# 好的做法
def good_func(items: Optional[list[int]] = None) -> list[int]:
    if items is None:
        items = []
    items.append(1)
    return items

print(good_func())  # [1]
print(good_func())  # [1] - 每次都是新的 list
print(good_func())  # [1]
````

````python
# 練習：為以下函式添加型別提示
def calculate_average(numbers):
    if not numbers:
        return None
    return sum(numbers) / len(numbers)

def merge_dicts(dict1, dict2):
    result = dict1.copy()
    result.update(dict2)
    return result

# 測試
# print(calculate_average([1, 2, 3, 4, 5]))
# print(merge_dicts({"a": 1}, {"b": 2}))
````

````python
# 解答
from typing import Any

def calculate_average(numbers: list[float]) -> float | None:
    if not numbers:
        return None
    return sum(numbers) / len(numbers)

def merge_dicts(
    dict1: dict[str, Any],
    dict2: dict[str, Any]
) -> dict[str, Any]:
    result = dict1.copy()
    result.update(dict2)
    return result

# 測試
print(calculate_average([1, 2, 3, 4, 5]))
print(calculate_average([]))
print(merge_dicts({"a": 1}, {"b": 2}))
````

### 練習 2：泛型函式

實作一個泛型的 `find_first` 函式。

````python
# 練習：找到列表中第一個符合條件的元素，若無則回傳 None
def find_first(items, predicate):
    for item in items:
        if predicate(item):
            return item
    return None

# 測試
# print(find_first([1, 2, 3, 4, 5], lambda x: x > 3))
````

````python
# 解答
from typing import TypeVar, Callable, Optional

T = TypeVar("T")

def find_first(
    items: list[T],
    predicate: Callable[[T], bool]
) -> Optional[T]:
    for item in items:
        if predicate(item):
            return item
    return None

# 測試
print(find_first([1, 2, 3, 4, 5], lambda x: x > 3))  # 4
print(find_first(["apple", "banana", "cherry"], lambda s: s.startswith("b")))  # banana
print(find_first([1, 2, 3], lambda x: x > 10))  # None
````

### 練習 3：Protocol 定義

定義一個 `Comparable` Protocol，讓任何可比較的物件都能使用。

````python
# 練習：定義 Comparable Protocol
# 你的程式碼
````

````python
# 解答
from typing import Protocol, TypeVar

class Comparable(Protocol):
    def __lt__(self, other: "Comparable") -> bool: ...
    def __gt__(self, other: "Comparable") -> bool: ...
    def __eq__(self, other: object) -> bool: ...

C = TypeVar("C", bound=Comparable)

def find_max(items: list[C]) -> C | None:
    if not items:
        return None
    return max(items)

# 測試
print(find_max([3, 1, 4, 1, 5, 9]))  # 9
print(find_max(["apple", "banana", "cherry"]))  # cherry
print(find_max([]))  # None
````

### 練習 4：TypedDict 使用

定義一個 API 回應的型別。

````python
# 練習：定義 ApiResponse TypedDict
# 你的程式碼
````

````python
# 解答
from typing import TypedDict, NotRequired, Any

class ApiResponse(TypedDict):
    status: int
    message: str
    data: NotRequired[dict[str, Any]]
    errors: NotRequired[list[str]]

def handle_response(response: ApiResponse) -> None:
    print(f"Status: {response['status']}")
    print(f"Message: {response['message']}")
    if "data" in response:
        print(f"Data: {response['data']}")
    if "errors" in response:
        for error in response["errors"]:
            print(f"Error: {error}")

# 測試
success_response: ApiResponse = {
    "status": 200,
    "message": "Success",
    "data": {"user_id": 1, "name": "Alice"}
}

error_response: ApiResponse = {
    "status": 400,
    "message": "Bad Request",
    "errors": ["Missing required field: email", "Invalid age format"]
}

print("=== Success Response ===")
handle_response(success_response)
print("\n=== Error Response ===")
handle_response(error_response)
````

## 小結

### TypeScript vs Python 型別提示對照表

| 概念 | TypeScript | Python |
|------|------------|--------|
| 基本型別 | `string`, `number`, `boolean` | `str`, `int`, `float`, `bool` |
| 陣列 | `string[]` 或 `Array<string>` | `list[str]` |
| 物件/字典 | `{ [key: string]: number }` | `dict[str, int]` |
| 元組 | `[string, number]` | `tuple[str, int]` |
| 可選 | `string \| null` | `Optional[str]` 或 `str \| None` |
| 聯合型別 | `string \| number` | `Union[str, int]` 或 `str \| int` |
| 函式型別 | `(x: number) => string` | `Callable[[int], str]` |
| 泛型 | `<T>` | `TypeVar("T")` |
| 介面 | `interface` | `Protocol` 或 `TypedDict` |
| 字面值型別 | `"a" \| "b" \| "c"` | `Literal["a", "b", "c"]` |
| 任意型別 | `any` | `Any` |
| 未知型別 | `unknown` | `object` |
| 常數 | `const` / `as const` | `Final` |
| 型別別名 | `type X = ...` | `TypeAlias` 或 `type X = ...` (3.12+) |
| 重載 | 函式重載 | `@overload` |

### 重點回顧

1. **型別提示是可選的**：Python 不強制使用型別提示，但使用它能提高程式碼品質
2. **執行時不檢查**：型別提示在執行時不會進行驗證，需要使用工具如 mypy 或 pyright
3. **Python 3.9+ 簡化語法**：可直接使用 `list[str]` 而非 `List[str]`
4. **Python 3.10+ Union 語法**：可使用 `X | Y` 而非 `Union[X, Y]`
5. **Protocol 實現鴨子型別**：不需要顯式繼承，只要結構相符即可
6. **TypedDict 定義字典結構**：類似 TypeScript 的 interface
7. **使用型別檢查工具**：mypy 和 pyright 是主流選擇
